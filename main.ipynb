```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def kmens_plot_3(df,labels):
    kmeans = KMeans(n_clusters=3, init='k-means++',n_init=10
                   )

    Y = kmeans.fit_predict(df)

# # Get the cluster centers
# centers = kmeans.cluster_centers_

# Plot the data points with their cluster assignments and cluster centers
    plt.scatter(df[Y==0,0],df[Y==0,1], s=10 , color='blue' ,label="Cluster 1")
    plt.scatter(df[Y==1,0],df[Y==1,1], s=10 , color='red' ,label="Cluster 2")
    plt.scatter(df[Y==2,0],df[Y==2,1], s=10 , color='green' ,label="Cluster 3")

    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    plt.title('K-means Clustering')
    plt.legend()
    plt.show()
```

```python
# import module
from geopy.geocoders import Nominatim
# initialize Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")
# Latitude & Longitude input
Latitude = "28.588303"
Longitude = "77.217697"

location = geolocator.reverse(Latitude+","+Longitude)

# Display
print(location)
```

```python
address = location.raw['address']
print('state name is ', address['state'])
print(address)
```

```python
def get_state_name(latitude,longitude):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.reverse(latitude+","+longitude)
    address = location.raw['address']
    return address['state']
```

```python
data['latitude']=data['latitude'].astype(str)
data['longitude']=data['longitude'].astype(str)
```

```python
data['State'] = data.apply(lambda row: get_state_name(row['latitude'], row['longitude']), axis=1)
```

```python
data['State'].value_counts()
```

```python
state_counts = data['State'].value_counts()

plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Number of Charging station in each State')
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()
```

```python
level_1_criteria = 7  # Charging stations with power capacity <= 7 kW are Level 1 chargers
level_2_criteria = 50  # Charging stations with power capacity > 7 kW and <= 50 kW are Level 2 chargers

# Create a function to categorize charging levels based on power capacity
def categorize_charging_level(power):
    if power == '15 kW':
        return "Level 1"
    elif power == '10(3.3 kW each)':
        return "Level 2"
    elif power == '142kW':
        return "Fast Charging"
```

```python
data['Charging Level'] = data['power'].apply(categorize_charging_level)
```
